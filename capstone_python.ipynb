#Exploratory Data Analysis (EDA) and Visualization

##Campaign Performance

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x= 'Acquisition_Cost',y = 'ROI' ,hue = 'Acquisition_Cost',data = df, )
plt.title("Relationship between Acquisition_Cost and roi")
plt.show()
```

```python
sns.barplot(x= 'Channel_Used',y= 'Conversion_Rate',hue = 'Campaign_Type',data = df )
plt.title("conversion rate for diffrent channel used")
plt.show()
```

```python
sns.boxplot(x= 'Engagement_Score',y = 'Campaign_Type',data = df)
plt.title("distribution of Engagement_Score different Campaign_Type")
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x= 'Company',y ='ROI',data = df)
plt.title("Average ROI by company")
plt.show()
```

```python
correlationmatrix = df[['Engagement_Score','Conversion_Rate']].corr()
```

```python
sns.heatmap(correlationmatrix,annot = True,cmap='icefire')
plt.title("corelation matrix")
plt.show()
```

#Customer Segmentation

```python
sns.countplot(x= 'Target_Audience', data = df)
plt.title('distribution of Target_Audience')
plt.show()
```

```python
plt.figure(figsize= (12,6))
sns.barplot(x ='Customer_Segment',y = 'Conversion_Rate', hue = 'Language' , data = df )
plt.title('Highest Conversion_Rate for eachLanguage')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Acquisition_Cost', y ='Customer_Segment',hue = 'Channel_Used',data = df)
plt.title('Distribution of Acquisition Cost Across Customer Segments by Channel Used')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x = 'Language',y = 'Conversion_Rate',hue = 'Language',data = df)
plt.title('Average Conversion_Rate by Language')
plt.show()
```

#Channel Effectiveness

```python
plt.figure(figsize = (11,6))
sns.barplot(x= 'Channel_Used',y = 'Engagement_Score',hue= 'Campaign_Type',data = df)
plt.title('the Engagement_Score for different Channels_Used')
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
channel_roi = df.groupby('Channel_Used')['ROI'].sum()
```

```python
plt.pie(channel_roi, labels=channel_roi.index, autopct='%1.1f%%')
plt.title('Distribution of Total ROI Across Channels')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x= 'Clicks',y='Impressions',hue='Channel_Used',data = df)
plt.title('Relationship between Clicks and Impressions')
plt.show()
```

#Time-Based Analysis

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Duration'], kde=True)
plt.title("disturbution of duration")
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Conversion_Rate', hue='Company', data=df,errorbar =None)
plt.title('conversion rate over a time')
plt.show()
```

```python
sns.lineplot(x='Engagement_Score',y= 'Date',data = df,errorbar = None)
plt.title('Trend of Engagement_Score over Date')
plt.show()
```

#Geographic Analysis

```python
plt.figure(figsize = (12,6))
sns.barplot(x='Location',y = 'Acquisition_Cost', data = df ,estimator= sum,color ='skyblue' )
plt.title("Acquisition cost by location")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Location',y='Conversion_Rate',hue='Target_Audience',data = df,palette='Set2')
plt.title("conversion rate by location ")
plt.show()
```

```python
Location_roi = df.groupby('Location')['ROI'].sum()
```

```python
plt.pie(Location_roi,labels=Location_roi.index,autopct='%1.1f%%')
plt.title('Distribution of total roi across locations')
plt.show()
```
